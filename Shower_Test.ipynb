```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import datetime
```

```python
with open('Shower_Test_Tags/tags 3.txt', 'r') as file:
    reader = csv.reader(file)

    database3 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false
    reached_24 = False

    for l in reader:

        line = l[0]
         
        if len(line)==24:                # If it is an EPC
            database3[line] = [[],[]]
            recent_EPC = line
        
        elif line == '[':
            data = []

        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database3[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database3[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23:
                reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ':
            pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass

with open('Shower_Test_Tags/tags 4.txt', 'r') as file:
    reader = csv.reader(file)

    database4 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false

    reached_24 = False

    for l in reader:

        line = l[0]
         
        if len(line)==24:                # If it is an EPC
            database4[line] = [[],[]]
            recent_EPC = line
        
        elif line == '[':
            data = []

        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database4[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database4[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23:
                reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ':
            pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass

with open('Shower_Test_Tags/tags 5.txt', 'r') as file:
    reader = csv.reader(file)

    database5 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false

    reached_24 = False

    for l in reader:

        line = l[0]
         
        if len(line)==24:                # If it is an EPC
            database5[line] = [[],[]]
            recent_EPC = line
        
        elif line == '[':
            data = []

        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database5[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database5[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23: reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ': pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass

with open('Shower_Test_Tags/tags 6.txt', 'r') as file:
    reader = csv.reader(file)

    database6 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false

    reached_24 = False

    for l in reader:

        line = l[0]
         
        if len(line)==24:                # If it is an EPC
            database6[line] = [[],[]]
            recent_EPC = line
        
        elif line == '[':
            data = []

        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database6[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database6[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23:
                reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ':
            pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass

with open('Shower_Test_Tags/tags 7.txt', 'r') as file:
    reader = csv.reader(file)

    database7 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false

    reached_24 = False

    for l in reader:

        line = l[0]
         
        if len(line)==24:                # If it is an EPC
            database7[line] = [[],[]]
            recent_EPC = line
        
        elif line == '[':
            data = []

        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database7[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database7[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23:
                reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ':
            pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass

with open('Shower_Test_Tags/tags 8.txt', 'r') as file:
    reader = csv.reader(file)
    database8 = dict()
    recent_EPC = ''
    Time_Or_Data = True                  # Time is true, Data is false
    reached_24 = False
    for l in reader:
        line = l[0] 
        if len(line)==24:                # If it is an EPC
            database8[line] = [[],[]]
            recent_EPC = line
        elif line == '[':
            data = []
        elif line == ']':                # End of list
            if len(data)!=0:
                if Time_Or_Data:
                    database8[recent_EPC][0] = data
                    Time_Or_Data = False
                else:
                    database8[recent_EPC][1] = data
                    Time_Or_Data = True
            data = []

        elif len(line)==8:               # If it is a time stamp
            h,m,s = line.split(':')

            if int(h)==23:
                reached_24 = True

            if reached_24 and int(h) < 12:
                day=2
            else:
                day=1

            data.append(datetime.datetime(1900, 1, day,int(h),int(m),int(s)))
        
        elif line ==' ': pass

        else:                            # If it is a data point
            try:
                data.append(float(line))
            except:
                pass
```

```python
import matplotlib.dates as mdates
myFmt = mdates.DateFormatter('%H:%M:%S')

fig, ax = plt.subplots(figsize=(14, 9.5))
ax.xaxis.set_major_formatter(myFmt)

for d in [database3, database4, database5, database6, database7, database8]:
    for k,v in d.items():
        if len(v[1]) > 1:
            ax.plot(v[0], v[1], label=k[-4:], linewidth=1.25, marker=".")

ax.set_ylabel("Temperature (°C)", fontsize=16.5)
ax.set_xlabel("Time (H:M:S)", fontsize=16.5)
ax.set_title("tags 3 through 8 data, Temperatures over Time", fontsize=16.5)
ax.set_ylim(10.0, 40.0)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc="upper right")
```
